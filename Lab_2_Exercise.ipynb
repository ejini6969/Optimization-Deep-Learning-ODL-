# Exercise

Using the boston_pre data, perform the data preprocessing

```python
import pandas as pd
import numpy as np

path = "/content/drive/MyDrive/boston_pre.csv"
boston = pd.read_csv(path)
```

```python
boston.head()
```

```python
boston.info()
```

```python
# drop rows if all contain null value
bost1 = boston.dropna(how = 'all')
bost1
```

```python
# drop all columns if only contain NULL / NAN value (Price2)
bost1 = bost1.dropna(axis = 1, how = 'all')
bost1
```

```python
bost1 = bost1.drop(["CHAS"], axis = 1) # drop CHAS column since all contain 0
bost1
```

```python
bost1.isnull().sum()
```

```python
# importing the class called SimpleImputer from impute model in sklearn
from sklearn.impute import SimpleImputer

# to replace the missing value we create below object for SimpleImputer class
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(bost1.iloc[:,:-1])

# replacing the missing value using transform method
bost1.iloc[:,:-1] = imputer.transform(bost1.iloc[:,:-1]) # do not include last column since all are valid value
```

```python
# convert age column from float to int type
bost1 = bost1.astype({'AGE':'int'})
bost1
```
