# Step 4: Taking care of the missing data

```python
# Total missing values for each feature / column
df.isnull().sum()
```

```python
#  drop rows where all cells in that row is NA
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
df1 = df.dropna(how='all') # only drop rows where all column entry is empty / null, by default: 'any'
df1
```

```python
df2 = df.dropna(how='any') # drop rows if it contains any missing / null value
df2
# since salary2 consists of all empty values, all rows will be dropped
```

```python
# drop column if they only contain missing values
df1 = df1.dropna(axis = 1, how = 'all')
df1
```

# Step 5: missing values

```python
# importing the class called SimpleImputer from impute model in sklearn
from sklearn.impute import SimpleImputer

# to replace the missing value we create below object for SimpleImputer class
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(df1.iloc[:,1:3])

# replacing the missing value using transform method
df1.iloc[:,1:3] = imputer.transform(df1.iloc[:,1:3])
```

```python
df1 = df1.astype({'Age': 'int', 'Salary': 'int'}) # convert columns from flaot to integer
df1
```

```python
df2 = df1.copy() #copy dataframe
```

# Step 6: Categorical Features

```python
df1['Marital'].value_counts() # count occurences of each category in column
```

```python
# label marital column
from sklearn.preprocessing import LabelEncoder
l1 = LabelEncoder()
l1.fit(df1['Marital'])
df1.Marital = l1.transform(df1.Marital)
print(df1) # 0 -> divorced, 1 -> married. 2 -> single
```

```python
# label purchased column
from sklearn.preprocessing import LabelEncoder
l1 = LabelEncoder()
l1.fit(df1['Purchased'])
df1.Purchased = l1.transform(df1.Purchased)
print(df1) # 0 -> No, 1 -> Yes
```

# Step 7: Split data into Y and X

```python
Y =df1["Purchased"]
X =df1.drop("Purchased",axis=1)
```

# Step 8: One Hot Encoding

```python
# for linear regression purposes (convert column values to categorical results)

# ColumnTransformer class of sklearn.compose can be used for transforming multiple categorical features
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])], remainder = 'passthrough')
df2_new = np.array(ct.fit_transform(df2))
df2_new
```

```python
df2_new = pd.DataFrame(df2_new, columns = ['Married', 'Divorced', 'Single', 'Age', 'Salary', 'Marital', 'Purchased'])
df2_new
```
